#Stock Market Trend Analysis and Financial Insights



#1.Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

sns.set(style="whitegrid")
```

#2.Load Dataset

```python
df = pd.read_csv("AAPL.csv")

print("Dataset Loaded Successfully")

df.head()
```

#3.Dataset Overview

```python
print("Dataset Shape:", df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
display(df.describe())
```

#4.Data Cleaning

Convert Date column

```python
df['Date'] = pd.to_datetime(df['Date'])
```

Sort data

```python
df = df.sort_values('Date')
```

#5.Statistical Analysis

```python
print("Average Closing Price:", df['Close'].mean())

print("Highest Price:", df['High'].max())

print("Lowest Price:", df['Low'].min())

print("Average Trading Volume:", df['Volume'].mean())
```

#6.Visualization

```python
#Stock Price Trend
plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['Close'])

plt.title("Stock Closing Price Trend")

plt.xlabel("Date")

plt.ylabel("Close Price")

plt.show()
```

```python
#High vs Low Price
plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['High'], label="High")

plt.plot(df['Date'], df['Low'], label="Low")

plt.legend()

plt.title("High vs Low Price Trend")

plt.show()
```

```python
#Trading Volume Trend
plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['Volume'])

plt.title("Trading Volume Over Time")

plt.xlabel("Date")

plt.ylabel("Volume")

plt.show()
```

```python
#Price Distribution
plt.figure(figsize=(8,5))

sns.histplot(df['Close'], bins=30)

plt.title("Closing Price Distribution")

plt.show()
```

Daily Return Analysis

Create daily returns:


```python
df['daily_return'] = df['Close'].pct_change()
```

```python
#Plot distribution:

plt.figure(figsize=(8,5))

sns.histplot(df['daily_return'].dropna(), bins=50)

plt.title("Daily Return Distribution")

plt.show()
```

```python
#Moving Average Trend
df['MA50'] = df['Close'].rolling(window=50).mean()

plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['Close'], label="Close Price")

plt.plot(df['Date'], df['MA50'], label="50-Day Moving Average")

plt.legend()

plt.title("Moving Average Trend")

plt.show()
```

#Correlation Analysis

```python
plt.figure(figsize=(6,4))

sns.heatmap(df[['Open','High','Low','Close','Volume']].corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Stock Variable Correlation")

plt.show()
```

#Automated Financial Report

```python
print("===== STOCK MARKET ANALYSIS REPORT =====")

print("\nAverage Closing Price:", df['Close'].mean())

print("\nHighest Stock Price:", df['High'].max())

print("\nLowest Stock Price:", df['Low'].min())

print("\nHighest Trading Volume:", df['Volume'].max())
```
